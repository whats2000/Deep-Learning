```python
import pandas as pd

# Pre-Processed Version
X_train_raw = pd.read_csv('data/X_train.csv')
Y_train_raw = pd.read_csv('data/Y_train.csv')
X_test_raw = pd.read_csv('data/X_test.csv')
Y_test_raw = pd.read_csv('data/Y_test.csv')
```

```python
data = pd.concat([X_train_raw, Y_train_raw], axis=1)
data
```

```python
import matplotlib.pyplot as plt

# 假设data是包含你的数据的DataFrame或数组
plt.hist(data[data['label'] == 0]['age'], bins=20, alpha=0.5, color='blue', label='Label 0')
plt.hist(data[data['label'] == 1]['age'], bins=20, alpha=0.5, color='red', label='Label 1')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution by Label')
plt.legend()  # 添加图例
plt.show()
```

```python
data.boxplot(column='age', by='label')
plt.ylabel('Age')
plt.title('Age Distribution by Label')
plt.suptitle('')
plt.show()
```

```python
features_to_compare = ['fnlwgt', 'capital_gain', 'capital_loss', 'hours_per_week']

for feature in features_to_compare:
    plt.figure(figsize=(8, 4))
    plt.hist(data[data['label'] == 0][feature], bins=20, alpha=0.5, color='blue', label='Label 0')
    plt.hist(data[data['label'] == 1][feature], bins=20, alpha=0.5, color='red', label='Label 1')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.title(f'{feature} Distribution by Label')
    plt.legend()
    plt.show()
```

```python
import seaborn as sns

# 假设data是包含你的数据的DataFrame或数组
features_to_compare = ['fnlwgt', 'capital_gain', 'capital_loss', 'hours_per_week']

for feature in features_to_compare:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x='label', y=feature, data=data)
    plt.xlabel('Label')
    plt.ylabel(feature)
    plt.title(f'{feature} Distribution by Label')
    plt.show()
```
